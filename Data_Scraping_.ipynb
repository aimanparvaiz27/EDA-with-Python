**Web scraping with Python**

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'

# Let's check the status of web page
page = requests.get(url)
print(page)
```

```python
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
#soup.find('table', class_ = 'wikitable sortable sticky-header-multi sort-under')
table = soup.find_all('table')[0]
print(table)
```

```python
col_title = table.find_all('th', scope = 'col')
print(col_title)
```

```python
table_titles = [title.text.strip() for title in col_title]
print(table_titles)
```

```python
df = pd.DataFrame(columns = table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    # Ensure the length of individual_row_data matches df.columns
    individual_row_data = (individual_row_data + [None] * len(df.columns))[:len(df.columns)]

    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv('output.csv', index=False)
```
